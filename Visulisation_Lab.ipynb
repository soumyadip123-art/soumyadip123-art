```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    data = pd.read_csv('/content/youtube_data_2000.csv')
    print("\n✅ Data Loaded Successfully!\n")
except:
    print("❌ Error: File not found!")
    exit()
```

```python
# 📌 Step 3: Show Basic Data
print("🔹 First 5 Rows:\n")
print(data.head())

print("\n🔹 Dataset Info:\n")
print(data.info())
```

```python
# 📌 Step 4: Category-wise Analysis
grouped = data.groupby("category").mean(numeric_only=True).reset_index()

print("\n📊 Average Values by Category:\n")
print(grouped)
```

```python
# 📌 Step 5: Visualization

# 🔹 Views Bar Chart
plt.figure()
sns.barplot(x="category", y="views", data=grouped)
plt.title("Average Views by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# 🔹 Likes Bar Chart
plt.figure()
sns.barplot(x="category", y="likes", data=grouped)
plt.title("Average Likes by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# 🔹 Comments Bar Chart
plt.figure()
sns.barplot(x="category", y="comments", data=grouped)
plt.title("Average Comments by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# 🔹 Pie Chart
plt.figure()
plt.pie(grouped["views"], labels=grouped["category"], autopct="%1.1f%%")
plt.title("Popularity Share (Views)")
plt.show()
```

```python
# 🔹 Correlation Heatmap
plt.figure()
sns.heatmap(data.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# 📌 Step 6: Engagement Rate
data["engagement"] = (data["likes"] + data["comments"]) / data["views"]

engagement_data = data.groupby("category")["engagement"].mean().reset_index()

plt.figure()
sns.barplot(x="category", y="engagement", data=engagement_data)
plt.title("Engagement Rate by Category")
plt.xticks(rotation=45)
plt.show()

print("\n✅ Analysis Completed Successfully!")
```
